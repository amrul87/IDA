```python
import pandas as pd
```

```python
# function to read txt data files and convert them to proper csv files
# txtFile: input filename (including directory if applicable)
# csvFile: output filename (including directory if applicable)
# vtabchar: vertical tab character in the original file (to be replaced with newline command '\n')
# delim: delimiter character used in the original file (to be replaced with comma)
def txt2csv(txtFile, csvFile, vtabchar, delim):
    with open(txtFile, 'r') as file:
        data = file.read().replace(vtabchar, '\n').replace(delim, ',')
    with open(csvFile, 'w') as file:
        file.write(data)    
    return csvFile
```

```python
# read and convert all the txt data files to csv
K1DI2 = txt2csv('./Data/Komponente/Komponente_K1DI2.txt', './Data/Komponente/Komponente_K1DI2.csv', '	', '\\')
K2LE1 = txt2csv('./Data/Komponente/Komponente_K2LE1.txt', './Data/Komponente/Komponente_K2LE1.csv', '', 'II')
K2LE2 = txt2csv('./Data/Komponente/Komponente_K2LE2.txt', './Data/Komponente/Komponente_K2LE2.csv', '', '\\')
K2ST1 = txt2csv('./Data/Komponente/Komponente_K2ST1.txt', './Data/Komponente/Komponente_K2ST1.csv', '', '|')
K3AG2 = txt2csv('./Data/Komponente/Komponente_K3AG2.txt', './Data/Komponente/Komponente_K3AG2.csv', '', '\\')
K7    = txt2csv('./Data/Komponente/Komponente_K7.txt'   , './Data/Komponente/Komponente_K7.csv'   , '', '	')
```
